## Gráfica de Funciones

````python
vector = np.linspace(0, 2*np.pi, 100)
fn = np.sin(vector)

plt.plot(vector, fn, color="red")
plt.title("Función Sinusoidal")
plt.xlabel("x")
plt.ylabel("f(x) = sin(x)")
plt.grid()
````

## Múltiples Gráficas

````python
vector = np.linspace(0, 2*np.pi, 100)
fn_1 = np.sin(vector)

fn_2 = 10 * np.sin(vector)

fn_3 = np.sin(vector) + 3

plt.figure(figsize=(17,5))
# Primera gráfica
plt.subplot(1,3,1)
plt.plot(vector, fn_1, color="red")
plt.title("Función Sinusoidal")
plt.xlabel("x")
plt.ylabel("f(x) = sin(x)")
plt.grid()

# Segunda gráfica
plt.subplot(1,3,2)
plt.plot(vector, fn_2, color="blue")
plt.title("Función Sinusoidal")
plt.xlabel("x")
plt.ylabel("f(x) = 10 * sin(x)")
plt.grid()

# Tercera gráfica
plt.subplot(1,3,3)
plt.plot(vector, fn_3, color="green")
plt.title("Función Sinusoidal")
plt.xlabel("x")
plt.ylabel("f(x) = sin(x) + 3")
plt.grid()
````

## Tipos de Gráfica

### Scatter

````python
vector = np.linspace(0, 2*np.pi, 50)
fn = np.sin(vector)

plt.scatter(vector, fn, color="red")
plt.title("Función Sinusoidal")
plt.xlabel("x")
plt.ylabel("f(x) = sin(x)")
plt.grid()
````

````python
vector = np.linspace(0, 2*np.pi, 50)
fn = np.sin(vector)

plt.stem(vector, fn)
plt.title("Función Sinusoidal")
plt.xlabel("x")
plt.ylabel("f(x) = sin(x)")
plt.grid()
````

# 3. Cargar Datos

Anteriormente cargamos datos de audio usando la librería librosa, cargamos un entrenamiento de un modelo de lenguaje usando pandas, pero existen distintas formas de cargar datos de todo tipo. A continuación, se cargará un simple archivo de texto.

````python
# Especifica la ruta del archivo de texto
archivo_txt = '/content/python-course/Recursos/Q&A_Teoría_de_Circuitos.txt'

# Abre el archivo en modo de lectura ('r')
with open(archivo_txt, 'r') as archivo:
    # Lee todo el contenido del archivo y lo almacena en una variable
    contenido = archivo.read()
````

````python
contenido
````

Podemos leer el archivo línea por línea, podemos usar un bucle for:

````python
with open(archivo_txt, 'r') as archivo:
    for linea in archivo:
        # Procesa cada línea según tus necesidades
        print(linea.strip())  # strip() elimina espacios en blanco al inicio y al final de la línea
````

Recuerda que la lectura de archivos en Python depende de la codificación del archivo. Si el archivo utiliza una codificación específica, puedes especificarla al abrir el archivo. Por ejemplo, si el archivo utiliza UTF-8, puedes hacer:



```python
with open(archivo_txt, 'r', encoding='utf-8') as archivo:
    # Operaciones de lectura aquí

```



1. `Modo de escritura ('w')`:

  * Este modo abre el archivo para escritura. Si el archivo ya existe, se truncará (eliminará su contenido anterior). Si el archivo no existe, se creará.

2. `Modo de anexado ('a')`:

  * Este modo abre el archivo para escritura, pero agrega nuevo contenido al final del archivo sin truncarlo.

3. `Modo de escritura y lectura ('w+')`:

  * Este modo abre el archivo tanto para escritura como para lectura.

4. `Codificación (encoding='utf-8')`

  * Podemos especificar la codificación utilizada.

````python
ruta_archivo = '/content/Q&A_Teoría_de_Circuitos.txt'

with open(ruta_archivo, 'w') as archivo:
    archivo.write("Esto es una prueba!")
````

````python
ruta_archivo = '/content/Q&A_Teoría_de_Circuitos.txt'

# Abre el archivo en modo de anexado ('a')
with open(ruta_archivo, 'a') as archivo:
    # Agrega el nuevo contenido al final del archivo
    archivo.write("\nEste es el nuevo contenido que se agregará al final.") # En este ejemplo, \n se usa para agregar una nueva línea antes del contenido adicional.
````

````python
with open(archivo_txt, 'r') as archivo:
    for linea in archivo:
        # Procesa cada línea según tus necesidades
        print(linea.strip())  # strip() elimina espacios en blanco al inicio y al final de la línea
````
